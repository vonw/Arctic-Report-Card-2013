#Plots for comparison to SAT and Cloud Cover for the 2013 Arctic Report Card

###All the plots below were created using ECMWF reanalysis data.

###Python Initialization

```python
# ....Read in necessary Python modules.
from datetime import datetime
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import scipy.io as sci

# ....Load monthly mean values of surface radiation components from the ECMWF reanalyses.
era = sci.loadmat('/Users/vonw/data/ArcticReportCard/2013/era_rads_spatial_2.mat')
```

```python
# ....Create variables from the era variable.
mos    = era['era'][0][0][0]
years  = era['era'][0][0][1]
dlwm   = era['era'][0][0][2]
dswm   = era['era'][0][0][3]
nlwm   = era['era'][0][0][4]
nswm   = era['era'][0][0][5]
lon    = era['era'][0][0][6]
lat    = era['era'][0][0][7]
ulwm   = era['era'][0][0][8]
uswm   = era['era'][0][0][9]
net    = era['era'][0][0][10]
```

```python
# ....Indices on the monthly values (252 total values (=21 years * 12 months/year); last 6 values are NaNs).
jans = range(0,252,12)
febs = range(1,252,12)
mars = range(2,252,12)
aprs = range(3,252,12)
mays = range(4,252,12)
juns = range(5,252,12)
juls = range(6,252,12)
augs = range(7,252,12)
seps = range(8,252,12)
octs = range(9,252,12)
novs = range(10,252,12)
decs = range(11,252,12)
```

####The largest positive anomaly in net radiation for this time period is seen over the Barents Sea.  This can be traced to a position anomaly in the net shortwave radiation in May and June of 2013. ###SEASONAL ANOMALY PATTERNS - Net Radiation Anomalies for different seasons (Oct 2012-Jun 2013)

```python
# ....Set up map projection
#m   = Basemap(projection='npstere', lon_0=-105., lat_0=60., boundinglat=60., resolution='l',round=True)
m   = Basemap(projection='npstere', lon_0=180., lat_0=60., boundinglat=60., resolution='l',round=True)   # Jim Overland's lon orientation
x,y = m(*np.meshgrid(lon,lat))
```

```python
############################ SEASONAL NET RADIATION for October 2012 - June 2013 ############################################################
# ................
fig=figure(figsize=(12,12))
fig.suptitle('SEASONAL ECMWF NET RADIATION ANOMALIES for Oct 2012 - Jun 2013 (base period - 1992-2012)')

# ....Oct 2012 - Dec 2012
subplot(2,2,1)
CS=m.contourf(x,y,transpose(net[:,:,249:252].mean(axis=2)-net[:,:,[octs, novs, decs]].mean(axis=2).mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('Oct-Dec 2012')

# ....Jan 2013 - Mar 2013
subplot(2,2,2)
CS=m.contourf(x,y,transpose(net[:,:,252:255].mean(axis=2)-net[:,:,[jans, febs, mars]].mean(axis=2).mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('Jan-Mar 2013')

# ....Apr 2013 - May 2013
subplot(2,2,3)
CS=m.contourf(x,y,transpose(net[:,:,255:257].mean(axis=2)-net[:,:,[aprs, mays]].mean(axis=2).mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('Apr-May 2013')

# ....Jun 2013
subplot(2,2,4)
CS=m.contourf(x,y,transpose(net[:,:,257]-net[:,:,juns].mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('Jun 2013')
```

####These plots of net radiation do NOT show a positive anomaly in net radiation in the autumn months, as is shown in the SAT plots in the ARC 2013.  However, the strong positive anomalies that begin to develop over the Laptev Sea in Apr/May 2013, and then cover most of the eastern Arctic Ocean, do appear to be consistent with the SAT anomalies.

###CLOUD COVER COMPARISON - Net Radiation Anomaly for January and June 2013

####These plots are intended to be compared to Jeff Key's plots of the Cloud Cover anomalies derived from MODIS data for January and June 2013.  These anomalies are relative to the corresponding monthly means over the period 2002 - 2010.

```python
# ....Set up map projection
#m   = Basemap(projection='npstere', lon_0=-105., lat_0=60., boundinglat=60., resolution='l',round=True)
#m   = Basemap(projection='npstere', lon_0=180., lat_0=60., boundinglat=60., resolution='l',round=True)   # Jim Overland's lon orientation
m   = Basemap(projection='npstere', lon_0=0., lat_0=60., boundinglat=60., resolution='l',round=True)   # Jeff Key's lon orientation
x,y = m(*np.meshgrid(lon,lat))
```

```python
############################ NET RADIATION for January 2013 and June 2013 ############################################################
# ................
fig=figure(figsize=(12,6))
fig.suptitle('ECMWF NET RADIATION ANOMALIES for January 2013 and June 2013 (base period - 2002-2010)')

# ....January 2013
subplot(1,2,1)
CS=m.contourf(x,y,transpose(net[:,:,252]-net[:,:,jans[11:21]].mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('January 2013')

# ....June 2013
subplot(1,2,2)
CS=m.contourf(x,y,transpose(net[:,:,257]-net[:,:,juns[11:21]].mean(axis=2)),range(-50,51,1),cmap=plt.cm.RdBu_r,extend='both')
CB=plt.colorbar(ticks=range(-50,51,10), shrink=0.8)
m.drawcoastlines()
m.drawparallels(range(60,90,10))
m.drawmeridians(range(-180,180,10))
plt.title('June 2013')
```

####In January, the net radiation is derived from the downwelling longware minus the upwelling longwave (since there is no solar radiation).  Therefore, the negative net radiation anomalies seen across the Arctic Ocean are driven by stronger than normal upwelling longwave, which is likely due to decreased cloud cover.

####In June, there are very strong positive net radiation anomalies over the Barents and Laptev Seas and, in general, over much of the eastern Arctic Ocean.  (There is also a strong anomaly over the northern Canadian continent.)  These are driven by strong positive anomalies in the balance of shortware radiation, which could be due to anomalously low cloud cover and/or sea ice concentrations.  The western Arctic Ocean (north of Canada) is characterized by weaker negatively anomalies.
