#Summary of ground-based measurements of the Surface Radiation Budget for the 2013 Arctic Report Card

####The plot below is a summary of measurements of the surface radiation budget from various sites around the Arctic Basin.  Each of these sites is a member institution of the [International Arctic Systems for Observing the Atmosphere](http://iasoa.org) (IASOA) group.  These data were kindly provided by researchers from each field station.

####The plot below was created by Dr. Chris Cox from CIRES.  It shows monthly averages (solid lines) of the complete records of the SRB components at each of six sites across the Arctic, plus the monthly averages for the first nine months of 2013 (where data are available).  The upper panel shows the downwelling components (shortwave and longwave), and the bottom panel shows upwelling components; the components that are zero during the winter months (December and January) are the shortwave components.
